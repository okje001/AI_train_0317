```python
#Exam_BinaryClassification_cancer , 유방암 분류 데이터
import sklearn
datasets = sklearn.datasets.load_breast_cancer()
print(datasets.keys())
x_data = datasets["data"]
y_data = datasets["target"]
print(datasets["target_names"])# malignant(악성) benign(양성)
features = datasets["feature_names"]
print(features)
# mean radius 평균 반지름
# mean texture 표면 질감
# mean perimeter 평균 경계
# mean area 평균 크기 .....
print(x_data[0])
print(y_data[0])
print(x_data[0][0])
```

```python
#데이터 분할(훈련데이터 70%, 테스트 데이터 15%, 훈련시검증데이터 15%
x_train,x_test,y_train,y_test = sklearn.model_selection.train_test_split(x_data,y_data,test_size=0.3,random_state=111,\
                                        stratify=y_data)#stratify 해당파일을 기준으로 골고루 분산해서 분할
x_valid,x_test,y_valid,y_test = sklearn.model_selection.train_test_split(x_test,y_test,test_size=0.5,random_state=111,\
                                        stratify=y_test)#stratify 해당파일을 기준으로 골고루 분산해서 분할
print("훈련데이터:",x_train.shape)
print("검증데이터:",x_valid.shape)
print("테스트데이터:",x_test.shape)
print("훈련정답:",y_train.shape)
print("검증정답:",y_valid.shape)
print("테스트정답:",y_test.shape)
```

```python
import numpy as np
# def dataConv2(x_conv):
#     for ix in range(len(x_conv[0])):
#         x_conv[:,ix]=(x_conv[:,ix]-np.mean(x_conv[:,ix]))/np.std(x_conv[:,ix])
#     return x_conv
def dataConvert(x_conv):
    scaler = sklearn.preprocessing.StandardScaler()
    x_conv=scaler.fit_transform(x_conv)
    return x_conv
# x_res = dataConv2(np.copy(x_train))
x_train = dataConvert(x_train)    
x_valid = dataConvert(x_valid)    
x_test = dataConvert(x_test)
# print(x_res[10])
print(x_train[10])
print(x_valid[10])
print(x_test[10])
import tensorflow as tf
from tensorflow.keras import Sequential, Input
from tensorflow.keras.layers import Dense
model = Sequential()
model.add(Input((x_train.shape[1],)))
model.add(Dense(1,activation="sigmoid"))
model.compile(loss="binary_crossentropy",optimizer="SGD",metrics=["acc"])
```

```python
fhist = model.fit(x_train,y_train,validation_data=(x_valid,y_valid),epochs=100)
```

```python
print(fhist.history.keys())
import matplotlib.pyplot as plt
plt.subplot(1,2,1)
plt.plot(fhist.history["loss"],label="train_loss")
plt.plot(fhist.history["val_loss"],label="valid_loss")
plt.legend()
plt.title("LOSSES")
plt.subplot(1,2,2)
plt.plot(fhist.history["acc"],label="train_acc")
plt.plot(fhist.history["val_acc"],label="valid_acc")
plt.legend()
plt.title("VALIDATION")
plt.show()
```

```python
res = model.evaluate(x_test,y_test)
print(res)
print("손실도 : ",res[0]," 정확률 : ",int(res[1]*10000)/100,"%")
# 3/3 [==============================] - 0s 2ms/step - loss: 132.6767 - acc: 0.8488
# [132.67674255371094, 0.8488371968269348]
# 손실도 :  132.67674255371094  정확률 :  84.88 %
```

```python
r = np.random.randint(0,len(x_test),1)
print(r)
y_pred = model.predict([x_test[r]])
print("악성종양" if y_pred[0][0]>0.5 else "양성종양")
print("악성종양" if y_test[r]>0.5 else "양성종양")
```

```python
#힛트맵
y_pred = model.predict(x_test)

y_pred = y_pred>0.5
y_pred=y_pred.astype("int")
y_test = y_test.reshape(len(y_test),-1)
print(y_pred.shape)
print(y_test.shape)
print(y_pred[:5])
print(y_test[:5])
import seaborn as sn
cm = sklearn.metrics.confusion_matrix(y_test,y_pred)
sn.heatmap(cm,annot=True)
plt.show()
```
